# Part 2. Text Analysis

## 2.1 extract high requency words in description

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.probability import *
from itertools import chain
#from tqdm import tqdm
import codecs
print("Import complete!")
```

```python
with open('stopwords.txt') as f:
    stop_words = f.read().splitlines()
stop_words = set(stop_words)
```

```python
# write your code here
# define your tokenize
tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?") 

description_data = final_wine_data_dropped["description"]
raw_data = []
for val in description_data:
    raw_data.append(val)
raw_data = ' '.join(raw_data)

def tokenizerRawData(raw_data):
    tokens = tokenizer.tokenize(raw_data)
    tokens = [token.lower() for token in tokens]
    tokens_filtered = [token for token in tokens if token not in stop_words]
    return tokens_filtered

print("Tokenizer created successfully!")
```

```python
# find top common words with document frequencies > 5000
# you may use function FreqDist() and sort()
most_common_words = FreqDist(sorted(tokenizerRawData(raw_data)))
results = []

for words, frequency in sorted(most_common_words.most_common()):
    if frequency > 5000:
        results.append((words, frequency))

freq_words = pd.DataFrame(results,columns=['words','frequency'])
freq_words
```

```python
# save your table to 'top_common_words.txt'  
f1 = open('HighFreq.txt', 'w')
f2 = open('top_common_words.txt', 'w')

for words, frequency in sorted(most_common_words.most_common()):
    if frequency > 5000:
        f1.write(words+"\n")
        f2.write(words+"\n")
f1.close()
f2.close()
print("Saved Successfully!")
```

## 2.2 Find key words for describing Shiraz using TF-IDF

```python
# select 'description' from 'variety' eqaul to  'Shiraz' 
final_wine_data_shiraz = final_wine_data_dropped[final_wine_data_dropped['variety'] == 'Shiraz']
shiraz_series = final_wine_data_shiraz['description']
```

```python
# use TfidfVectorizer to calculate TF-IDF score
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(analyzer='word', stop_words = 'english')

raw_data1 = []
for val in shiraz_series:
    raw_data1.append(tokenizerRawData(val));

response = tfidf.fit_transform([' '.join(value) for value in raw_data1])
response.shape
```

```python
# find words with TF-IDF score >0.4 and sort them
vocab = tfidf.get_feature_names()
shiraz_tfidf_words = []

for word, weight in zip(vocab, response.toarray()[0]):
    if(weight > 0.4):
        shiraz_tfidf_words.append((word,weight))
        
shiraz_tfidf_words_df = pd.DataFrame(shiraz_tfidf_words, columns=['word','weight'])
#displaying the table for words in shiraz variety description having tfidf score above 0
shiraz_tfidf_words_df
```

```python
# save your table to 'Shirazkey.txt'   
f3 = open('key_Shiraz.txt', 'w')
for word, weight in sorted(zip(vocab, response.toarray()[0])):
    if(weight > 0.4):
        f3.write(word + "\n")
f3.close()
```
